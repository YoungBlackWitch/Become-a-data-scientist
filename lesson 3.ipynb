```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
outerfunc = lambda x: x**2
```

```python
x = np.linspace(-5,5,101)
z = lambda x: x**3
```

```python
# 1. Напишите функцию, которая на вход принимает x и функцию в качестве аргументов и рисует график

def some_fun(x, outerfunc, save_name=None):
    y = outerfunc(x)
    plt.plot(x, y)
    plt.show()
```

```python
some_fun(x, outerfunc)
```

```python
# 2. Добавьте аргумент, дающий возможность сохранения графика в файл

def some_fun(x, outerfunc, save_name=None):
    y = outerfunc(x)
    plt.plot(x, y)

    if save_name:
        plt.savefig(save_name)
        plt.close()
    else:
        plt.show()
```

```python
some_fun(x, outerfunc, 'figure_1.png')
```

```python
title_dict = {'fontsize':20,
              'fontweight':'bold',
              'color':'#808080',
             'family':'Times New Roman'}

labels_dict = {'fontsize':14,
              'fontweight':'bold',
              'color':'#808080',
             'family':'Times New Roman'}
```

```python
style ={
    'title' : {
        'name':'График', 'font': title_dict
    },
    'xlabel':{
        'name':'Xlebel', 'font': labels_dict
    },
    'ylabel':{
        'name':'Ylebel', 'font': labels_dict
    }
}
```

```python
# 3. Добавьте аргументы, позволяющие сделать график более симпатичным

def some_fun(x, outerfunc, clr, save_name=None, style_params=None):
    y = outerfunc(x)
    plt.plot(x, y, color=clr)
    plt.grid(color='lightgrey')

    if style_params:
        plt.title(style_params.get('title').get('name'), fontdict=style_params.get('title').get('font'))
        plt.xlabel(style_params.get('xlabel').get('name'), fontdict=style_params.get('xlabel').get('font'))
        plt.ylabel(style_params.get('ylabel').get('name'), fontdict=style_params.get('ylabel').get('font'))
    if save_name:
        plt.savefig(save_name)
        plt.close()
    else:
        plt.show()
```

```python
some_fun(x, outerfunc, 'green', style_params=style)
```

```python
some_fun(x, z, 'red', style_params=style)
```
